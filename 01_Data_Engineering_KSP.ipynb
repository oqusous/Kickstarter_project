```python
# four columns contain dictionary looking data- these are 'profile', 'category', 'creator' and 'location'
# however, they are str type. The following syntax is to transform them into dict type.

# check number of items that are str
types = []
for i in df.index:
    types.append(type(df.at[i, 'category']))
```

```python
# transformer

def transformer(df,column_name):
    i_list = []
    for i in df.index:
        try:
            if isinstance(df.at[i, column_name], str):
                df.at[i, column_name] = json.loads(df.at[i, column_name])
        except:
            print(i) #print index of which rows did not convert
            i_list.append(i) # append to a list to inspect
            continue
    return i_list
```
